```python
from sklearn.preprocessing import LabelEncoder
```

```python
len1 = LabelEncoder()
len2 = LabelEncoder()
len3 = LabelEncoder()
len4 = LabelEncoder()
len5 = LabelEncoder()
len6 = LabelEncoder()
len7 = LabelEncoder()
len8 = LabelEncoder()
len9 = LabelEncoder()

trdata['country'] = len1.fit_transform(trdata['country'])
trdata['bank_account'] = len2.fit_transform(trdata['bank_account'])
trdata['location_type'] = len3.fit_transform(trdata['location_type'])
trdata['cellphone_access'] = len4.fit_transform(trdata['cellphone_access'])
trdata['gender_of_respondent'] = len5.fit_transform(trdata['gender_of_respondent'])
trdata['marital_status'] = len6.fit_transform(trdata['marital_status'])
trdata['relationship_with_head'] = len7.fit_transform(trdata['relationship_with_head'])
trdata['education_level'] = len8.fit_transform(trdata['education_level'])
trdata['job_type'] = len9.fit_transform(trdata['job_type'])
```

```python
gender_count = trdata2['gender_of_respondent'].value_counts()
# Create pie chart
fig, ax = plt.subplots()
ax.pie(gender_count, labels=gender_count.index, autopct='%1.1f%%')
ax.set_title('Gender Distribution')
plt.show()
```

```python
age_groups = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['0-17', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']
# Create age groups based on the age column
trdata2['age_groups'] = pd.cut(trdata2['age_of_respondent'], bins=age_groups, labels=labels)
# Create a histogram of age groups
fig, ax = plt.subplots(figsize=(8, 6))
trdata2['age_groups'].value_counts().sort_index().plot(kind='bar', color='purple', ax=ax)
ax.set_xlabel('Age Group')
ax.set_ylabel('Count')
ax.set_title('Customer Age Group Distribution')
# Add labels to histogram bars
for i in ax.containers:
  ax.bar_label(i)

plt.show()
```

```python
c = pd.crosstab(trdata2["bank_account"],trdata2["country"])
withbank = np.array(c.iloc[1:].values)
withoutbank = np.array(c.iloc[0:1].values)
countries = ['Kenya', 'Rwanda', 'Tanzania', 'Uganda']
with_bank = np.reshape(withbank, -1)
without_bank =  np.reshape(withoutbank, -1)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
axes[0, 0].pie([with_bank[0], without_bank[0]], labels=['With Bank', 'Without Bank'], autopct='%1.1f%%')
axes[0, 1].pie([with_bank[1], without_bank[1]], labels=['With Bank', 'Without Bank'], autopct='%1.1f%%')
axes[1, 0].pie([with_bank[2], without_bank[2]], labels=['With Bank', 'Without Bank'], autopct='%1.1f%%')
axes[1, 1].pie([with_bank[3], without_bank[3]], labels=['With Bank', 'Without Bank'], autopct='%1.1f%%')

axes[0, 0].set_title(countries[0])
axes[0, 1].set_title(countries[1])
axes[1, 0].set_title(countries[2])
axes[1, 1].set_title(countries[3])

plt.subplots_adjust(wspace=0.4, hspace=0.4)

plt.show()
```
